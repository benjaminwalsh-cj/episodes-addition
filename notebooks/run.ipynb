```python
def get_npi_idx(npi):
    idx_tuple = np.where(manual_check_npis == npi)

    idx = idx_tuple[0][0]
    return idx
```

```python
idx = get_npi_idx(1558369397)
sum(all_egm_shap[0][idx])
```

```python
def compare_shap_values(npi: int, class_idx: int) -> list:
    '''Generate key comparison values across all vs no egm for the provided NPI
    Args:
        npi (`int`): NPI to check
        class_idx (`int`): index of the category to compare
    Returns:
        List of comparison results
    '''

    idx = get_npi_idx(npi=npi)

    all_egm_shap_values = all_egm_shap[class_idx][idx]

    no_egm_shap_values = no_egm_shap[class_idx][idx]

    all_egm_sum = sum(all_egm_shap_values)
    no_egm_sum = sum(no_egm_shap_values)

    shap_differences = all_egm_shap_values[:8834] - no_egm_shap_values

    return shap_differences
    
```

```python
shap_differences = compare_shap_values(1558369397, 0)
```
